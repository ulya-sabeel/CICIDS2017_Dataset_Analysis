```python
#importing pandas library
import pandas as pd 
  
data=pd.read_csv("data.csv")
 
data.head(5)
```

```python
#convert data from dataframe to matrix
data1=data.values
#Split data into training data (X) and target label(Y)
X=data1[:,:-1] #gets elements of array but the last one
Y=data1[:,-1]
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

check_occurence = [i for i in Y if i == 1]
print("total length of test data= ",len(Y))
print("Attack = ",len(check_occurence))
print("Benign = ", (len(Y) - len(check_occurence)))

b = len(Y) - len(check_occurence)
a = len(check_occurence)

label = ['Benign', 'Attack']
samples = [b, a]
dpi = mpl.rcParamsDefault['figure.dpi']
mpl.rcParams['figure.dpi'] = dpi * 1.5

# this is for plotting purpose
index = np.arange(len(label))
plt.bar(index, samples)
plt.xlabel('Target Labels', fontsize=10)
plt.ylabel('Samples', fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=0)
plt.title('Attack and Benign data samples in the data')
plt.show()
```
