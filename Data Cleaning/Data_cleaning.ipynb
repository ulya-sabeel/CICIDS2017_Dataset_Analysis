```python
import os
import pandas as pd 

# load preprocessed dataset 
data=pd.read_csv("raw_attack_data.csv")

##check if data_train has both attack (1) and normal (0) flows 
print(data['Label'].value_counts())
```

```python
data_new=data.replace(to_replace ="No Label",value =1) 
```

```python
data_new.head(1)
```

```python
# iterating the columns with their indices
for idx ,col in enumerate(data_new.columns): 
    print(str(idx) + " " + str(col))
```

```python
# dropping passed columns 
data_new.drop(data_new.columns[[0,1,2,3,4,5,6,40]], axis = 1, inplace = True) 
data_new.head(2)
```

```python
#drop nan values
import numpy as np
final_data= data_new.replace([np.inf, -np.inf, "Infinity"], np.nan).dropna(axis=0, how ='any')
# comparing sizes of data frames 
print("Old data frame length:", len(data), "\nNew data frame length:",len(final_data), "\nNumber of rows with at least 1 NA value: ",(len(data)-len(final_data))) 
```

```python
tempo=final_data
#replace negatives with zeros
tempo[tempo < float(0)] = 0
tempo.to_csv('cleandata.csv',index=False, header=True)
```
